```python
import time
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import numpy as np 
import re
import datetime
```

```python
chrome_options = Options()
chrome_options.add_argument("--headless")
chrome_driver = "C:\webdrivers\chromedriver.exe" #give your webdriver path
driver = webdriver.Chrome(chrome_driver)

driver.maximize_window()
site_url = "https://uae.sharafdg.com/"
product_name = "Gaming Consoles"   #change the product
search_path = "//*[@id='woocommerce-product-search-field']"
total_count = "//*[@id='stats']"

# def product_search(site_url,product_name,search_path,total_count):
#     driver = webdriver.Chrome("C:\webdrivers\chromedriver.exe")
driver.get(site_url)

    # Searching products
search_bar = driver.find_element_by_xpath(search_path)
search_bar.clear()
search_bar.send_keys(product_name)
search_bar.send_keys(Keys.ENTER)
    
time.sleep(40)
count = driver.find_element_by_xpath(total_count).text
# return count
print(count)
```

```python
SCROLL_PAUSE_TIME = 2

# Get scroll height
last_height = driver.execute_script("return document.body.scrollHeight")

while True:
    # Scroll down to bottom
    driver.execute_script("window.scrollTo(0, 1000*document.body.scrollHeight);")

    # Wait to load page
    time.sleep(SCROLL_PAUSE_TIME)

    # Calculate new scroll height and compare with last scroll height
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height
```

```python
value = re.findall("\d+",count)
image_url_lst=[]
for j in value:
    for y in range(1,int(j)):
        xpath = "//*[@id='hits']/div["
        final_xpath = xpath + str(y) + "]/div[1]/a"
        image_url = driver.find_element_by_xpath(final_xpath).get_attribute("href")
        image_url_lst.append(image_url)
```

```python
category_code = 'CA035' #check the code & change
catalogue_code = 'CT005' #check the code & changechange
vendor_code = 'VC004' #sharafdg
catalogue_name = "Consoles"

final_products_list = []
for i in image_url_lst:
    driver.get(i)
    category_code = category_code
    catalogue_code = catalogue_code
    vendor_code= vendor_code
    
    #product_code
    try:
        ProductCode = 'NA'
    except:
        ProductCode = 'NA'
        
    #vendor_name
#     try:
#         VendorCode = 'uae.sharafdg.com'
#     except:
#         VendorCode = 'uae.sharafdg.com'
        
    #sku_code
    try:
        SKU = driver.find_element_by_css_selector('div.mainproduct-info > div > div.product-title > div:nth-child(4) > p > span').text
    except:
        SKU = 'NA'
        
    #product_name    
    try:
        ProductName = driver.find_element_by_css_selector('div.mainproduct-info > div > div.product-title > h1').text
    except:
        ProductName = 'NA'
    
    #product_desc
    try:
        aa = driver.find_element_by_css_selector('#prodcut-tab1 > div').text
        ProductDesc = re.sub('\n',',',aa)
    except:
        ProductDesc = 'NA'
        
    #url    
    try:
        URL = i
    except:
        URL = 'NA'
    
    #image
    try:
        MainImage = driver.find_element_by_css_selector('div.mainproduct-slider.clearfix.slick-initialized.slick-slider > div > div > div.mp-item.slick-slide.slick-current.slick-active > div > img').get_attribute('src')
    except:
        MainImage = 'NA'
    
    #catalogue_name
    try:
        CatalogueName = driver.find_element_by_css_selector('div.breadcrumb-wrapper > div > div > ul > li:nth-child(2) > a > span').text
    except:
        CatalogueName = 'NA'
        
    #category_name
    try:
        CategoryName = catalogue_name
    except:
        CategoryName = catalogue_name
        
    #brand_code    
    try:
        BrandCode = 'NA'
    except:
        BrandCode = 'NA'
        
    #brand_name   
    try:
        BrandName ='NA'
    except:
        BrandName = 'NA'
    
    #sub_brand_name
    try:
        SubBrandName ='NA'
    except:
        SubBrandName = 'NA'
    
    #stock_availability
    try:
        StockAvailablity = 'Yes'
    except:
        StockAvailablity = 'NA'
        
    #offer_price    
    try:
        offer_price1 = driver.find_element_by_css_selector('div > div.product-price > div.price').text
        Offer = float(offer_price1.replace('AED',''))
    except:
        Offer = 'NA'
    
    #regular_price
    try:
        regular_price1 =driver.find_element_by_css_selector('div > div.product-price > div.cross-price').text
        RegularPrice = float(regular_price1.replace('AED',''))
    except:
        RegularPrice = 'NA'
    
    #model_name
    try:
        ModelName = 'NA'
    except:
        ModelName = 'NA'
        
    #model_num    
    try:
        ModelNumber = driver.find_element_by_css_selector('div.product-title > div:nth-child(3) > p > span').text
    except:
        ModelNumber = 'NA'
        
    #rating
    try:
        rating_value1 = driver.find_element_by_css_selector('div.mainproduct-info > div > div.product-title > div.in-block.rating-badge > div > div.rating-details.rating-details-block.in-block.prod-extra > span').text
        a = rating_value1
        b = a.replace('(','')
        RatingValue = b.replace(')','')    
    except:
        RatingValue = 'NA'
    
    #best_rating
    try:
        BestRating1 = driver.find_element_by_css_selector('div.mainproduct-info > div > div.product-title > div.in-block.rating-badge > div > div.rating-details.rating-details-block.in-block.prod-extra > span').text
        a = BestRating1
        b = a.replace('(','')
        BestRating = b.replace(')','') 
    except:
        BestRating= 'NA'
    
    #review_count
    try:
        ReviewCount = driver.find_element_by_css_selector('div.product-title > div.in-block.rating-badge > div > div.review-details.in-block.reviewCount > span').text
    except:
        ReviewCount = 'NA'
        
    #sub_category_name    
#     try:
#         sub_category_name = driver.find_element_by_css_selector('div.breadcrumb-wrapper > div > div > ul > li:nth-child(3) > a > span').text
#     except:
#         sub_category_name = 'NA'
    
    #date_inserted    
    now = datetime.datetime.now()
    DateInserted = now.strftime("%d%m%Y")
    
    #date_updated
    now = datetime.datetime.now()
    DateUpdated = now.strftime("%d%m%Y")

    prod_dict = {
        'ProductCode': ProductCode,
        'CatalogueCode': catalogue_code,
        'CategoryCode': category_code,
        'VendorCode': vendor_code,
        'SKU': SKU,
        'ProductName': ProductName,
        'ProductDesc': ProductDesc,
        'URL' : URL,
        'MainImage': MainImage,
        'CatalogueName': CatalogueName,
        'CategoryName': CategoryName,
        'BrandCode' : BrandCode,
        'BrandName': BrandName,
        'SubBrandName': SubBrandName,
        'StockAvailablity': StockAvailablity,
        'Offer/SalePrice' : Offer,
        'RegularPrice': RegularPrice,
        'ModelName': ModelName,
        'ModelNumber': ModelNumber,
        'RatingValue': RatingValue,
        'BestRating/Rating': BestRating,
        'ReviewCount': ReviewCount,
        'DateInserted':DateInserted,
        'DateUpdated':DateUpdated
    }
    final_products_list.append(prod_dict)

```
